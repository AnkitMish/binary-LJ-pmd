```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

def mdplot(dir=".."):
    plt.figure();
    fig, axes = plt.subplots(figsize=(16,8),nrows=2, ncols=2);

    msd = pd.read_csv(dir+'/msd.d', names=['time','msd','count'])
    vac = pd.read_csv(dir+'/vac.d', names=['time','vac','count'])
    gr = pd.read_csv(dir+'/gr.d', names=['r','g(r)','N(r)','count'])

    msd.plot(marker='o', ax=axes[0,0], x='time',y='msd')
    vac.plot(marker='o', ax=axes[0,1], x='time',y='vac')
    gr.plot(marker='o', ax=axes[1,0], x='r',y='g(r)')
    gr.plot(marker='o', ax=axes[1,1], xlim=[0,3], ylim=[0,20], x='r',y='N(r)')
    fig.savefig(dir+'/pmd.png',bbox_inches='tight')
```

```python
mdplot();
```
