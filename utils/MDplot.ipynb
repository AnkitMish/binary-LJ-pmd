```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

def mdplot(dir=".."):
    plt.figure();
    fig, axes = plt.subplots(figsize=(16,16),nrows=3, ncols=2);

    msd = pd.read_csv(dir+'/msd.d', names=['MDstep','msd','count'])
    vac = pd.read_csv(dir+'/vac.d', names=['MDstep','vac','count'])
    gr = pd.read_csv(dir+'/gr.d', names=['r','g(r)','N(r)','count'])
    sq = pd.read_csv(dir+'/sq.d', names=['q','S(q)'])
    ft = pd.read_csv(dir+'/ft.d', names=['w','F(w)'])
    
    msd.plot(marker='o', ax=axes[0,0], x='MDstep',y='msd')
    vac.plot(marker='o', ax=axes[0,1], x='MDstep',y='vac')
    gr.plot(marker='o', ax=axes[1,0], x='r',y='g(r)')
    gr.plot(marker='o', ax=axes[1,1], xlim=[0,3], ylim=[0,30], x='r',y='N(r)')
    sq.plot(marker='o', ax=axes[2,0], x='q',y='S(q)')
    ft.plot(marker='o', ax=axes[2,1], x='w',y='F(w)')

    fig.savefig(dir+'/pmd.png',bbox_inches='tight',dpi=150)
```

```python
mdplot();
```
